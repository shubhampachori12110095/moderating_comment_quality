```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
df_ted = pd.read_csv("TED Talks as of 10.17.2016 - TEDTalks by date.csv")
df_ted.head()
```

```python
df_ted['speaker_topic'] = df_ted.speaker_name + ' ' + df_ted.headline
```

```python
df_ted.head()
```

```python
yt = df_com_yt['title']
ted_choices = df_ted['speaker_topic']
matchlist = []
counter = 0
for ident in yt:
    top_match = process.extractOne(ident, ted_choices)
    alt_match = process.extract(ident, ted_choices, limit=2)
    matchdict = {"original": ident,
                 "top_match": top_match,
                 "alt_match": alt_match}
    matchlist.append(matchdict)
    counter += 1
    print(counter)
```

```python
with open('ted_desc_match.pkl','wb') as picklefile:
    pickle.dump(matchlist,picklefile)
```

```python
matchlist[:5]
```

```python
df_match = pd.DataFrame.from_dict(matchlist)
```

```python
df_match.head()
```

```python
alt = df_match.alt_match
alt_1 = [i[0] for i in alt]
alt_2 = [i[1] for i in alt]
```

```python
df_match['alt_1'] = alt_1
df_match['alt_2'] = alt_2
df_match.drop('alt_match',axis=1,inplace=True)
```

```python
df_match.to_csv('match_titles.csv')
```
